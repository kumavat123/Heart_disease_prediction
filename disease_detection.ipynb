```python
import pandas as pd
df=pd.read_csv('/kaggle/input/heart-disease-data/heart_disease_uci.csv')
```

```python
numeric_cols=df.select_dtypes(include=['number']).columns
df[numeric_cols]= df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df[numeric_cols].hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df[numeric_cols].corr(),annot=True,cmap='coolwarm')
plt.title('Numeric Feature Correlations')
plt.show()
```

```python
cat_cols=df.select_dtypes(include=['object']).columns.tolist()
if 'num' in cat_cols:
  cat_cols.remove('num')
```

```python
# to drop a row which have num 1=heart disease & 0=not heart Disease
x=df.drop('num',axis=1)
y=(df['num']>0).astype(int)
```

```python
#used encoding in this code ,print final feature colunms
x=pd.get_dummies(x,columns=cat_cols)
print("Final feature colunms: ",x.columns)
```

# Day 3 : training and testinng **bold text**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# training and testing to split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

new_value=(x-mean)/standard deviation

```python
# to put in the scaler values standard and resized
scaler=StandardScaler()
#learn and understand the mean and deviation
x_train_scaled=scaler.fit_transform(x_train)
# to transform in scaled value
x_test_scaled=scaler.transform(x_test)
```

```python
# it is about classification
from sklearn.linear_model import LogisticRegression
```

```python
# fiting the model
lr_model=LogisticRegression()  #giving admission to new student
lr_model.fit(x_train_scaled,y_train) # training step
```

## Model Evaluation

```python
# we are working on superwise learning
# work on model evaluation
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# after prediction
y_pred_lr=lr_model.predict(x_test_scaled)
print("Logistic Regression Accuracy: ",accuracy_score(y_test,y_pred_lr))
print("Classification Report:\n",classification_report(y_test,y_pred_lr))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred_lr))
```

# Accuracy score -75 to 98 considered to be good

# day :4 Random Forest and feature Importance

```python
# ploting the confusion matrix as TN TP FP FN
cm=confusion_matrix(y_test,y_pred_lr)
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(x_train_scaled,y_train)
y_pred_rf=rf_model.predict(x_test_scaled)
print("Random Forest Accuracy: ",accuracy_score(y_test,y_pred_rf))
print("Classification Report:\n",classification_report(y_test,y_pred_rf))
```

# feature importance

```python
feat_imp= pd.Series(rf_model.feature_importances_,index=x.columns)
feat_imp.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importance')
plt.show()
```

```python
import joblib
import pandas as pd

user_df = pd.read_csv('heart_dataset.csv')

# getting columns list from dataframe
numeric_cols= df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
bool_cols=df.select_dtypes(include='bool').columns.tolist()

#Dropping columns which are extra in user_df than required to avoid error
# we are checking that the columns which are present in the data are same or not
numeric_cols=[col for col in numeric_cols if col in user_df.columns]
cat_cols=[col for col in cat_cols if col in user_df.columns]
bool_cols=[col for col in bool_cols if col in user_df.columns]

# fill the missing numeric column & cat column
user_df[numeric_cols]=user_df[numeric_cols].fillna(user_df[numeric_cols].mean())

for col in cat_cols:
  user_df[col]=user_df[col].fillna('Unknown')
for col in bool_cols:
  user_df[col]=user_df[col].astype(int)

#One-hot encoding cat columns
user_df_encoded=pd.get_dummies(user_df,columns=cat_cols)

#Allign columns
user_df_encoded=user_df_encoded.reindex(columns=x.columns,fill_value=0)

# Scale data
scaler = joblib.load('heart_scaler.pkl')
user_df_scaled = scaler.transform(user_df_encoded)

# Load the trained model
model = joblib.load('heart_rf_model.pkl')
user_df_scaled = scaler.transform(user_df_encoded)

# prediction
model=joblib.load('heart_rf_model.pkl')
preds=model.predict(user_df_scaled)
user_df['Heart_Disease_Prediction']=preds

#show result
print(user_df)
```
